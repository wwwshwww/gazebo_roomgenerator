```python
import numpy as np
import quaternion
from shapely.geometry import Polygon

def vec_to_trans(vec):
    a = np.identity(len(vec)+1)
    a[:len(vec),len(vec)] = vec
    return a

def add_dimension(points, value=0):
    return np.concatenate([points[np.newaxis,:,i].T if i<len(points[0]) else np.full([len(points),1], value) for i in range(len(points[0])+1)], 1)
    
def get_square_horizon(base_pos, radius, z_angle=0):
    d1 = [1,1,-1,-1]
    d2 = [1,-1,-1,1]
    poss = np.array([np.quaternion(1,d1[i]*radius,d2[i]*radius,0) for i in range(4)])
    q = quaternion.from_euler_angles(0,0,z_angle)
    pq = q*poss*q.conj()
    
    return np.array([[p.x, p.y, p.z] for p in pq])+base_pos

def get_extended_face(face_vertices, length):
    face_bottom = np.copy(face_vertices)
    face_bottom[:,2] += length
    return np.concatenate([face_vertices, face_bottom])
    
def distance_filtered_poly(polygon, distance=0.01):
    c = np.round(polygon.exterior.coords, 3)
    ban = np.zeros([len(c)], dtype=int)
    for i in range(len(c)):
        if not ban[i]:
            n = np.linalg.norm(c-c[i], axis=1) < distance
            n[i] = False
            ban = np.logical_or(ban, n) 
    
    return Polygon(c[np.logical_not(ban)])

def unique_augmented_poly(polygon, length, center=None):
    p = np.array(polygon.exterior.coords)
    
    if center is None:
#         c = np.zeros([2])
        c = polygon.centroid
#         c = np.mean(p[:-1], axis=0)
    else:
        c = center

    r = p - c
    n = r / np.linalg.norm(r, axis=1, ord=2)[np.newaxis,:].T
    return Polygon(p+n*length).convex_hull
```

```python
import numpy as np
import trimesh
from shapely.geometry import Polygon

from pcg_gazebo.generators.creators import create_models_from_config
from pcg_gazebo.generators.shapes import random_rectangle, random_points_to_triangulation, random_rectangles


def create_wall_model(room_length_max, wall_height, wall_thickness, model_name='room', color='random'):
    config = [
        dict(
            type='extrude',
            args=dict(
                polygon=random_points_to_triangulation(
                    x_min=-room_length_max/2,
                    x_max=room_length_max/2,
                    y_min=-room_length_max/2,
                    y_max=room_length_max/2
                ),
                height=wall_height,
                thickness=wall_thickness,
                extrude_boundaries=True,
                name=model_name,
                color=color
            )
        )
    ]
    return create_models_from_config(config)

def create_box_models(sizes, model_name='box', color='white'):
    config = [
        dict(
            type='box_factory',
            args=dict(
                size=sizes,
                name=model_name,
                use_permutation=True,
                color=color
            )
        )
    ]
    return create_models_from_config(config)
    
def create_cube_models(length, count, model_name='cube', color='white'):
    sizes = [[length, length, length]]*count
    return create_box_models(sizes, model_name, color)

def get_mesh_from_model(model):
    col = model.get_link_by_name(model.name).collisions[0]
    col_mesh_before_trans = col.geometry._geometry_entity.get_meshes()[0]

    return col_mesh_before_trans, col.pose.position

def get_corrected_mesh_from_model(model):
    import copy
    
    mesh, t = get_mesh_from_model(model)
    c_mesh = copy.copy(mesh)
    c_mesh.apply_transform(vec_to_trans(t)) #-
    return c_mesh

def get_poly_from_model(model, mesh_slice_height): ## height; ROOM_WALL_HEIGHT/2
    c_mesh = get_corrected_mesh_from_model(model)
    poly = trimesh.path.polygons.projected(c_mesh, (0,0,mesh_slice_height))
    return poly

def get_interior_poly_from_extrude_mesh(mesh, mesh_slice_height):
    extrude_poly = trimesh.path.polygons.projected(mesh, (0,0,mesh_slice_height))
    return distance_filtered_poly(Polygon(extrude_poly.interiors[0]))

def get_interior_poly_from_extrude_model(model, mesh_slice_height):
    extrude_poly = get_poly_from_model(model, mesh_slice_height)
    extrude_interior_poly = distance_filtered_poly(Polygon(extrude_poly.interiors[0])) #-
    return extrude_interior_poly

def sample_from_poly(polygon, count):
    return trimesh.path.polygons.sample(polygon, count)

```

```python
from pcg_gazebo.task_manager import ProcessManager
import shapely
from shapely.geometry.polygon import Polygon
from sklearn.cluster import DBSCAN
import trimesh

class CubeRoomGenerator():
    
    def __init__(self,
                 obstacle_count,
                 obstacle_size=0.6,
                 agent_size=0.3,
                 target_size=0.2,
                 room_length_max=8.2,
                 room_wall_height=0.8,
                 room_wall_thickness=0.05,
                 wall_threshold=0.1,
                 ros_host="localhost", ros_port=11311, gazebo_host='localhost', gazebo_port=11345):
        
        target_sim = dict(
            ros_host=ros_host,
            ros_port=ros_port,
            gazebo_host=gazebo_host,
            gazebo_port=gazebo_port
        )
        
        self.process_manager = ProcessManager(**target_sim)
        self.gazebo_proxy = self.process_manager.get_gazebo_proxy()
        
        self.obstacle_count = obstacle_count
        self.obstacle_size = obstacle_size
        self.agent_size = agent_size
        self.target_size = target_size
        self.room_length_max = room_length_max
        self.room_wall_height = room_wall_height
        self.room_wall_thickness = room_wall_thickness
        self.wall_threshold = wall_threshold
        
        self.wall_model = None
        self.obstacle_models = list()
        self.wall_mesh = None
        self.wall_interior_poly = None
        self.obstacle_poss = list()
        self.obstacle_oris = list()
        
        self.prepare_to_gen_room()
        
        self.obstacle_label = list()
        self.target_models = list()
        self.zone_vertices_2d = list()
        self.zone_polys = list()
        self.zone_hulls = list()
        self.target_poss = list()
        
        self.prepare_to_gen_targets()
        
        
    def prepare_to_gen_room(self):
        self.wall_model, self.obstacle_models = self.generate_room_components()
        
        self.wall_mesh = get_corrected_mesh_from_model(self.wall_model) #-
        self.wall_interior_poly = get_interior_poly_from_extrude_mesh(self.wall_mesh, self.room_wall_height/2) #-
        
        self.obstacle_poss, self.obstacle_oris = self.generate_obstacles_pose() ## [[n,3], [n,3]]
        
    def prepare_to_gen_targets(self):
        ##### ---
        self.obstacle_label = self.get_obstacle_cluster(eps=np.sqrt((self.obstacle_size**2)*2)+self.agent_size/2)
        self.target_models = create_cube_models(self.target_size, max(self.obstacle_label)+1, 'target', 'cyan')
        
        cls_count = max(self.obstacle_label)+1
        
        self.zone_vertices_2d = [None]*cls_count
        self.zone_polys = [None]*cls_count
        for i in range(cls_count):
            pts = self.obstacle_poss[self.obstacle_label==i]
            rts = self.obstacle_oris[self.obstacle_label==i]
            vertices = [get_square_horizon(p, self.obstacle_size/2, r[2]) for p,r in zip(pts,rts)] #-
            vertices_all = np.concatenate(vertices)

            self.zone_vertices_2d[i] = trimesh.convex.hull_points(vertices_all[:,:2])
            self.zone_polys[i] = shapely.ops.unary_union([Polygon(v) for v in vertices])
            
        self.zone_hulls = [None]*cls_count
        for i in range(cls_count):
            p3 = np.zeros([len(self.zone_vertices_2d[i]),3])
            p3[:,:2] = self.zone_vertices_2d[i]
            self.zone_hulls[i] = trimesh.Trimesh(vertices=get_extended_face(p3, self.obstacle_size)).convex_hull
            
        self.target_poss = self.generate_target_pose()
            
    def generate_target_pose(self):
        zone_samples = [None]*len(self.zone_hulls)
        for i in range(len(zone_samples)):
            ## sample_surface return tuple, so choice 0 index
            samples = trimesh.sample.sample_surface_even(self.zone_hulls[i], 100)[0]
            zone_samples[i] = samples[
                np.logical_and(
                    np.not_equal(samples[:,2], 0), 
                    np.not_equal(samples[:,2], self.obstacle_size)
                )
            ]

        choiced_goal = np.zeros([len(self.zone_hulls),3])
        for i in range(len(choiced_goal)):
            choiced_goal[i] = zone_samples[i][np.random.choice(range(len(zone_samples[i])))]

        choiced_goal[:,2] = self.target_size/2
        
        return choiced_goal
            
    @property
    def map_poly(self):
        all_poly = []
        for polys in self.zone_polys:
            if isinstance(polys, Polygon):
                all_poly.append(polys)
            else:
                for p in polys:
                    all_poly.append(p)

        map_poly = Polygon(
            list(self.wall_interior_poly.exterior.coords), 
            map(lambda p: list(p.exterior.coords), all_poly)
        )
        return map_poly
        
    
    def generate_room_components(self):
        wall = create_wall_model(self.room_length_max, self.room_wall_height, self.room_wall_thickness)[0] #-
        obstacles = create_cube_models(self.obstacle_size, self.obstacle_count) #-
        return wall, obstacles
    
    def generate_obstacles_pose(self):
        pos = add_dimension(
            sample_from_poly(
                unique_augmented_poly(self.wall_interior_poly, -1*(self.wall_threshold+self.obstacle_size))
            , self.obstacle_count)
        , self.obstacle_size/2) #-
        
        ori = np.zeros_like(pos)
        ori[:,2] = np.random.random([len(pos)])*np.pi*2
        
        return pos, ori
    
    def generate_room(self):
        self.wall_model.spawn(
            gazebo_proxy=self.gazebo_proxy, 
            robot_namespace=self.wall_model.name,
            pos=[0,0,self.room_wall_height/2]
        )
        
        for model, p, r in zip(self.obstacle_models, self.obstacle_poss, self.obstacle_oris):
            model.spawn(
                gazebo_proxy=self.gazebo_proxy, 
                robot_namespace=model.name,
                pos=list(p),
                rot=list(r)
            )
            
    def generate_targets(self):
        for model, p in zip(self.target_models, self.target_poss):
            model.get_link_by_name('link').disable_collision()
            model.spawn(
                gazebo_proxy=self.gazebo_proxy, 
                robot_namespace=model.name,
                pos=list(p)
            )
        
    def get_obstacle_cluster(self, eps): ## eps; np.sqrt((OBSTACLE_SIZE**2)*2)+AGENT_SIZE/2
        db = DBSCAN(eps=eps, min_samples=1).fit(np.array(self.obstacle_poss))
        return db.labels_
    
    
    
```

```python
generator = CubeRoomGenerator(
    obstacle_count=15,
    obstacle_size=0.6,
    target_size=0.2,
    agent_size=0.3
)

generator.generate_room()
generator.generate_targets()
generator.map_poly
```
